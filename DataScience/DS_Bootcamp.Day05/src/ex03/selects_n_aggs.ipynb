## Selects and aggregations

```python
import pandas as pd
```

```python
df = pd.read_json("../ex02/auto.json")
df.set_index("CarNumber", inplace=True)
df
```

### Selects

```python
df_fines_high_2100 = df[df["Fines"] > 2100]
df_fines_high_2100
```

```python
df_fines_high_2100_2 = df[(df["Fines"] > 2100) & (df["Refund"] == 2)]
df_fines_high_2100_2  
```

```python
df_model = df[df["Model"].isin(["Focus", "Corolla"])]
df_model 
```

```python
df_car_number = df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
df_car_number
```

### Aggregations with the make and the model

### Make bo'yicha median Fines ni ko'rsatish

```python
df_median_fines_by_make = df.groupby('Make')['Fines'].median()
df_median_fines_by_make
```

### Make va Model bo'yicha median Fines ni ko'rsatish

```python
df_median_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].median()
df_median_fines_by_make_model
```

### Make va Model bo'yicha fines sonini ko'rsatish

```python
df_count_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].count()
df_count_fines_by_make_model
```

### Make va Model bo'yicha minimum va maximum fines ni ko'rsatish

```python
df_min_max_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
df_min_max_fines_by_make_model
```

### Make va Model bo'yicha fines ning standard deviation ni ko'rsatish

```python
df_std_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].std()
df_std_fines_by_make_model
```

### Aggregations with the car number

### CarNumber bo'yicha fines ni kamayish tartibida saralash

```python
df_fines_grouped_by_car_number = df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False)
df_fines_grouped_by_car_number
```

### Eng ko'p jarima to'lagan CarNumber ni tanlash

```python
top_car_number_1 = df_fines_grouped_by_car_number.idxmax()
df_top_car_number_1 = df[df.index == top_car_number_1]
```

```python
top_car_number_1
```

```python
df_top_car_number_1
```

### CarNumber bo'yicha fines summasi bo'yicha saralash

```python
df_fines_sum_by_car_number = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
df_fines_sum_by_car_number
```

### Eng qimmat jarima to'lagan CarNumber ni tanlash

```python
top_car_number_2 = df_fines_sum_by_car_number.idxmax()
df_top_car_number_2 = df[df.index == top_car_number_2]
```

```python
top_car_number_2
```

```python
df_top_car_number_2
```
