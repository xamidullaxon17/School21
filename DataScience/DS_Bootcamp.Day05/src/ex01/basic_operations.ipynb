## Basic operations

```python
import pandas as pd
```

```python
path = "../ex00/data/feed-views.log"
df = pd.read_csv(path, sep="\t", names=['datetime', 'user'])
df
```

### Extract the year, month, day, hour, minute, and second from the values of that column to the new columns

```python
df["datetime"] = pd.to_datetime(df["datetime"])
df['year'] = df["datetime"].dt.year
df['month'] = df["datetime"].dt.month
df['day'] = df["datetime"].dt.day
df['hour'] = df["datetime"].dt.hour
df['minute'] = df["datetime"].dt.minute
df['second'] = df["datetime"].dt.second
df
```

### Created new daytime column

```python
bins = [0, 4, 7, 11, 17, 20, 24]
labels = ["night", "early morning", "morning", "afternoon", "early evening", "evening"]
df["daytime"] = pd.cut(df["hour"], bins=bins, labels=labels, right=False, include_lowest=True)
df
```

### Set index

```python
df.set_index("user", inplace=True)
df
```

### calculate the number of elements

```python
total = df.count()
total
```

```python
daytime_counts = df["daytime"].value_counts()
daytime_counts
```

### sorted values by hour, minute, and second

```python
df_sorted = df.sort_values(by=["hour", "minute", "second"])
df_sorted
```

### Calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
min_hour = df["hour"].min()
max_hour = df["hour"].max()
daytime_mode = df["daytime"].mode()[0]
daytime_mode
```

### who visited the page at those hours

```python
max_night_hour = df[df["daytime"] == "night"]["hour"].max()
night_user = df[df["hour"] == max_night_hour].index[0]
night_user
```

```python
min_morning_hour = df[df["daytime"] == "morning"]["hour"].min()
morning_user = df[df["hour"] == min_morning_hour].index[0]
morning_user
```

### Users 3 earliest and latest hours in the morning

```python
morning_df = df[df["daytime"] == "morning"]
earliest_morning = morning_df.nsmallest(3, "hour")
earliest_morning["hour"]
```

```python
latest_morning = morning_df.nlargest(3, "hour")
latest_morning['hour']
```
